### Bar plot

```python
# Bar plot representation of the gender distribution
y_female=len(female)
y_male=len(male)
plt.bar("Male", y_male, color="b")
plt.bar( "Female", y_female, color="r")
plt.ylabel("Number of Students")
plt.title("Gender Distribution")
plt.show()
```

```python
# Bar plot representation of the education distribution
key=list(df.groupby("Education")["Education"].count().index)
value=list(df.groupby("Education")["Education"].count().values)
plt.bar(key, value, color= ["r", "b"])
plt.show()
```

### Box plot

```python
x_labels=["Age", "Weight", "Height"]
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,7))

y_female=[female["Age"], female["Weight"], female["Height"]]
ax1.boxplot(y_female)
ax1.set(title="Boxplot of Female Students")
ax1.set_xticklabels(x_labels)
ax1.set_yticks(np.arange(25,201,25))
ax1.grid()

y_male=[male["Age"], male["Weight"], male["Height"]]
ax2.boxplot(y_male)
ax2.set(title="Boxplot of Male Students")
ax2.set_xticklabels(x_labels)
ax2.grid()
```

### Scatter plot

```python
x_female=female["Weight"]
y_female=female["Height"]
x_male=male["Weight"]
y_male=male["Height"]
plt.figure(figsize=(8, 5), dpi=90)
plt.scatter(x_female, y_female, color="r", label="Female")
plt.scatter(x_male, y_male, color="b", label="Male")

plt.title("Attribute Values of Students")
plt.xlabel("Weight(kg)")
plt.ylabel("Height(cm)")
plt.legend()
plt.show()
```
